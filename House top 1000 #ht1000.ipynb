```python
import requests
page = requests.get("https://ht1000.nl/no_stem_2020/rt.php")
```

```python
page.status_code
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
htList = soup.find_all('li', class_='row')
htList
```

```python
import html
import re
artists = [re.sub('^\s*|\s*$', '', html.unescape(r.find_all('div', class_='col-xs-4')[0].get_text())) for r in htList]
artists
songs = [re.sub('^\s*|\s*$', '', html.unescape(r.find_all('div', class_='col-xs-4')[1].get_text())) for r in htList]
songs
```

```python
import pandas as pd
ht = pd.DataFrame({
    'Artist': artists,
    'Song': songs
})
ht
```

```python
ht.to_csv('ht1000.csv',index=False)
```
